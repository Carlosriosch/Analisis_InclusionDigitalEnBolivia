Análisis de una campaña de digitalización en la educación en Bolivia
estos son los datos de una encuesta final de los estudiantes 

```python
import pandas as pd
from pandas_ods_reader import read_ods
```

```python
estudiantes=pd.read_csv('estudiantes_fin.csv')     
# load a sheet based on its index (1 based)
sheet_idx = 1
col = read_ods('diccionario-de-variables-final-estudiantes.ods', sheet_idx)
```

```python
estudiantes.columns=col['Descripción']
colum=col['Descripción']
```
